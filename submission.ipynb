```python
import pandas as pd
import matplotlib.pyplot as plt

train=pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
number = LabelEncoder()
train['gender'] = number.fit_transform(train['gender'].astype('str'))
test['gender'] = number.fit_transform(test['gender'].astype('str'))
```

```python
product = {'A':1,'B':2, 'C': 3,'D':4,'E':5, 'F':6, 'G':7, 'H':8, 'I':9, 'J':10}
train.replace(product, inplace=True)
test.replace(product, inplace=True)
import numpy as np
train = train.replace(np.NaN, 0)
test = test.replace(np.NaN, 0)
```

```python
train=train.drop(['DateTime'], axis=1)
test=test.drop(['DateTime'], axis=1)

train.head(5)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```
